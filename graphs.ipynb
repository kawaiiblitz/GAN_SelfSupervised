```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have read the CSV into a pandas DataFrame called df
# df = pd.read_csv('your_file.csv')
baseline = pd.read_csv('./output/stats/stats_b_100p.csv')
frozen_9 = pd.read_csv('./frozen_weights/stats/stats_ft_j9_100p.csv')
ft_9 = pd.read_csv('./output/stats/stats_ft_j9_100p.csv')

# To rename the first column:
baseline.rename(columns={baseline.columns[0]: 'epoch'}, inplace=True)
frozen_9.rename(columns={frozen_9.columns[0]: 'epoch'}, inplace=True)
ft_9.rename(columns={ft_9.columns[0]: 'epoch'}, inplace=True)
```

```python
# Assume the CSV file has two columns: 'epoch' and 'accuracy'
# where 'epoch' is the epoch number and 'accuracy' is the accuracy for that epoch.
# This will plot the accuracy for each epoch.
fig = plt.figure(figsize=(15, 15))
axs1 = fig.subplots(ncols=3, nrows=3)
axs1[0,0].plot(baseline['epoch'], baseline['train_loss'], label='Training', color='lightseagreen' )
axs1[0,0].plot(baseline['epoch'], baseline['val_loss'], label='Validation', color='hotpink' )
axs1[0,0].set_title('Baseline Loss')
axs1[0,0].set_xlabel('Epoch')
axs1[0,0].set_ylabel('Accuracy')
axs1[0,0].set_ylim([0.4, 0.9])
axs1[0,0].legend()
axs1[0,0].grid()
axs1[0,1].plot(frozen_9['epoch'], frozen_9['train_loss'], label='Training', color='lightseagreen' )
axs1[0,1].plot(frozen_9['epoch'], frozen_9['val_loss'], label='Validation', color='hotpink' )
axs1[0,1].set_title('JigsawGAN Loss (frozen)')
axs1[0,1].set_xlabel('Epoch')
axs1[0,1].set_ylabel('Accuracy')
axs1[0,1].set_ylim([0.4, 0.9])
axs1[0,1].legend()
axs1[0,1].grid()
axs1[0,2].plot(ft_9['epoch'], ft_9['train_loss'], label='Training', color='lightseagreen' )
axs1[0,2].plot(ft_9['epoch'], ft_9['val_loss'], label='Validation', color='hotpink' )
axs1[0,2].set_title('JigsawGAN Loss (thawed)')
axs1[0,2].set_xlabel('Epoch')
axs1[0,2].set_ylabel('Accuracy')
axs1[0,2].set_ylim([0.4, 0.9])
axs1[0,2].legend()
axs1[0,2].grid()

axs1[1,0].plot(baseline['epoch'], baseline['train_accuracy'], label='Training', color='lightseagreen' )
axs1[1,0].plot(baseline['epoch'], baseline['val_accuracy'], label='Validation', color='hotpink' )
axs1[1,0].set_title('Baseline Accuracy')
axs1[1,0].set_xlabel('Epoch')
axs1[1,0].set_ylabel('Accuracy')
axs1[1,0].set_ylim([0.6, 1])
axs1[1,0].legend()
axs1[1,0].grid()
axs1[1,1].plot(frozen_9['epoch'], frozen_9['train_accuracy'], label='Training', color='lightseagreen' )
axs1[1,1].plot(frozen_9['epoch'], frozen_9['val_accuracy'], label='Validation', color='hotpink' )
axs1[1,1].set_title('JigsawGAN Accuracy (frozen)')
axs1[1,1].set_xlabel('Epoch')
axs1[1,1].set_ylabel('Accuracy')
axs1[1,1].set_ylim([0.6, 1])
axs1[1,1].legend()
axs1[1,1].grid()
axs1[1,2].plot(ft_9['epoch'], ft_9['train_accuracy'], label='Training', color='lightseagreen' )
axs1[1,2].plot(ft_9['epoch'], ft_9['val_accuracy'], label='Validation', color='hotpink' )
axs1[1,2].set_title('JigsawGAN Accuracy (thawed)')
axs1[1,2].set_xlabel('Epoch')
axs1[1,2].set_ylabel('Accuracy')
axs1[1,2].set_ylim([0.6, 1])
axs1[1,2].legend()
axs1[1,2].grid()

axs1[2,0].plot(baseline['epoch'], baseline['train_iou'], label='Training', color='lightseagreen' )
axs1[2,0].plot(baseline['epoch'], baseline['val_iou'], label='Validation', color='hotpink' )
axs1[2,0].set_title('Baseline IoU')
axs1[2,0].set_xlabel('Epoch')
axs1[2,0].set_ylabel('Accuracy')
axs1[2,0].set_ylim([0.3, 0.9])
axs1[2,0].legend()
axs1[2,0].grid()
axs1[2,1].plot(frozen_9['epoch'], frozen_9['train_iou'], label='Training', color='lightseagreen' )
axs1[2,1].plot(frozen_9['epoch'], frozen_9['val_iou'], label='Validation', color='hotpink' )
axs1[2,1].set_title('JigsawGAN IoU (frozen)')
axs1[2,1].set_xlabel('Epoch')
axs1[2,1].set_ylabel('Accuracy')
axs1[2,1].set_ylim([0.3, 0.9])
axs1[2,1].legend()
axs1[2,1].grid()
axs1[2,2].plot(ft_9['epoch'], ft_9['train_iou'], label='Training', color='lightseagreen' )
axs1[2,2].plot(ft_9['epoch'], ft_9['val_iou'], label='Validation', color='hotpink' )
axs1[2,2].set_title('JigsawGAN IoU (thawed)')
axs1[2,2].set_xlabel('Epoch')
axs1[2,2].set_ylabel('Accuracy')
axs1[2,2].set_ylim([0.3, 0.9])
axs1[2,2].legend()
axs1[2,2].grid()

plt.show()
```
